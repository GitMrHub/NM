```python
import numpy as np

e = 0.0001
A = np.array([[864, 307, 269, -369, 275],[307, 537, 195, -280, 192],[269, 195, 412, -289, -4],\
              [-369, -280, -289, 375, -174],[275, 192, -4, -174, 377]])
```

```python
y = np.array([1,1,1,1,1])
x = y/y.max()
y = np.dot(A, x)
I_k = np.average(y/x)
I = 0
x = y/y.max()
k=1
while abs(I_k - I) >=e:
    k+=1
    I = I_k
    y = np.dot(A, x)
    I_k = y/x
    x = y/y.max()
    print('\n')
    print(k, 'ітерація: y=', y)
    print('y(i)k/y(i)k-1=',  I_k)
    I_k = np.average(I_k)
print("\nнайбільше власне число", I_k)
print('відповідний власний вектор', x)
```

```python
def get_max(B):
    matr = abs(B)
    for i in range(len(matr)):
        matr[i, i] = 0
        max_el = matr.max()
        ind = np.unravel_index(np.argmax(matr, axis=None), matr.shape)
    return max_el, ind
```

```python
def summ(B):
    summ=0
    for i in range(len(B)):
        for j in range(len(B[0])):
            if i!=j:
                summ+=B[i,j]**2
    return summ
```

```python
max_el, ind = get_max(A)
A_new = A.copy()
i = ind[1] 
j = ind[0]
k=1
vect = np.eye(A_new.shape[0])
while  summ(A_new) >= e:
    T = (A_new[i,i] - A_new[j,j])/(2*A_new[i,j])
    if T==0:
        c = np.sqrt(2)/2 
        s = np.sqrt(2)/2
    elif T<0:
        t = 1/(T - np.sqrt(T**2 + 1))
        c = 1/(np.sqrt(1+ t**2))
        s = c*t
    else:
        t = 1/(T + np.sqrt(T**2 + 1))
        c = 1/(np.sqrt(1+ t**2))
        s = c*t
    k+=1
    print('\n\nk =', k)
    print('c, s =', c, s)
    print('c^2 + s^2 = ', c**2 + s**2)
    T_matr = np.eye(A_new.shape[0])
    T_matr[i,i] = c
    T_matr[j,j] = c
    T_matr[i,j] = -s
    T_matr[j,i] = s
    vect = np.dot(vect, T_matr)
    B = np.dot(np.dot(T_matr.T, A_new), T_matr)
    max_el, ind = get_max(B)
    A_new = B
    print('A =', A_new)
    print('max_element =', max_el)
    print('index = ',ind)
    sum_d =0
    sum_nd = 0
    for i in range(len(B)):
        for j in range(len(B)):
            if j==i:
                sum_d += B[i,j]**2
            else:
                sum_nd+= B[i,j]**2
    print('pow sum of diag elem =', sum_d)
    print('pow sum of non-diag elem =', sum_nd)
    print('full sum =', sum_nd + sum_d)
    i = ind[1] 
    j = ind[0]
for i in range(5):
    print('\nвласне число',A_new[i,i])
    print('власний вектор',vect[:,i]/np.max(vect[:,i]))
```
