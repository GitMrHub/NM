```python
h= 0.05
while h<=0.2:
    print('h =', h)
    x0 = 0
    y0 = 2/3
    func = lambda x,y :2*x*(x-y)/(1+x**2)
    while (round(x0 + h, 3) <=1):
        n_1 = func(x0, y0)
        n_2 = func(x0+h/2, y0+h*n_1/2)
        n_3 = func(x0+h/2, y0+h*n_2/2)
        n_4 = func(x0+h, y0+h*n_3)
        y0+=h/6*(n_1 + 2*n_2 + 2*n_3 + n_4)
        x0+=h
        print('y({0}) = {1}'.format(round(x0, 2), round(y0, 9))) 
    h*=2
```
