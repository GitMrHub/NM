```python
import numpy as np 
   
x = np.arange(0, 1.2, 0.2) 
y = np.flip(np.arange(0, 1.2, 0.2)) 
arr = np.zeros((6, 6)) 
n = len(arr)
eps = 0.001
f = np.full((n-2, n-2), False) 

for i in range(0, n-1):
    arr[0][i] = 50 -10*i
    
arr[:,i+1] = 0
arr[:, 0] = 50*y**2 
arr[i+1] = 60*x*(1-x**2)

for i in range(1, n-1): 
        arr[i] = np.array([arr[i][0] + (arr[i][-1] - arr[i][0])/(n-1)*k for k in range(n)]) 
#print(arr)
k = 0 
while not f.all(): 
    k += 1 
    for i in range(1, n-1): 
        for j in range(1, n-1): 
            new_x = 1/4*(arr[i-1][j] + arr[i+1][j] + arr[i][j-1] + arr[i][j+1]) 
            if abs(new_x - arr[i][j]) <= eps: 
                f[i-1][j-1] = True 
            arr[i][j] = new_x 
    print(k) 
    print(np.round(arr, 4), '\n') 
```
